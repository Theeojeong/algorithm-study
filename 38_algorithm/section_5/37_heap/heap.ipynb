1. 현재 인덱스 * 2 -> 왼쪽 자식의 인덱스
2. 현재 인덱스 * 2 + 1 -> 오른쪽 자식의 인덱스 1 * 2 + 1 =3
3. 현재 인덱스 // 2 -> 부모의 인덱스

```python
from logging import root


class MaxHeap:
    def __init__(self):
        self.items = [None]

    def insert(self, value):
        self.items.append(value)
        current_index = len(self.items) - 1

        while current_index != 1:
            parent_index = current_index // 2
            if self.items[current_index] > self.items[parent_index]:
                self.items[current_index], self.items[parent_index] = self.items[parent_index], self.items[current_index] 
                current_index = parent_index
            else:
                break
    # 1. root와 맨 끝의 위치를 바꾼다
    # 2. 그리고 바뀐 root_node 를 자식과 비교한다
    # 3. 그리고 자식들이 크다면, 둘 중 더 큰 자식과 위치를 바꾼다
    # 4. 이 과정을 바닥을 찍을 때까지 혹은 내가 자식보다는 클때까지 반복한다
    # 5. 그리고 나서 1번의 root 노드를 반환한다
    def delete(self):
        self.items[1], self.items[-1] = self.items[-1], self.items[1]
        prev_max = self.items.pop()
        #              m
        #              c
        #                    
        #    0   1  2  3  4  5
        # [None, 6, 3, 1, 2, 3]
        current_index = 1
        while current_index <= len(self.items) - 1:
            left_child_index = current_index * 2
            right_child_index = current_index * 2 + 1
            max_index = current_index

            if (
                left_child_index < len(self.items) - 1
                and self.items[max_index] < self.items[left_child_index]
            ):
                max_index = left_child_index
            if (
                right_child_index < len(self.items) - 1
                and self.items[max_index] < self.items[right_child_index]
            ):
                max_index = right_child_index

            if max_index == current_index:
                break

            self.items[max_index], self.items[current_index] = (
                self.items[current_index],
                self.items[max_index],
            )
            current_index = max_index

        return prev_max
```
