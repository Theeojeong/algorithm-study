# 📊 그래프(Graph) 완벽 정리

## 1. 그래프란 무엇인가? 🎯
그래프는 **'점'**과 그 점들을 잇는 **'선'**으로 이루어진 자료구조입니다.

| 용어 | 설명 | 예시 |
|---|---|---|
| 정점 (Vertex / Node) | 데이터가 담기는 지점 | 서울, 부산, 사용자 A |
| 간선 (Edge) | 점들 사이의 관계나 연결 통로 | 고속도로, 친구 관계 |

💡 **실생활 예시**
- 지하철 노선도: 역(정점) + 철로(간선)
- SNS: 사용자(정점) + 팔로우/친구 관계(간선)
- 인터넷: 웹페이지(정점) + 하이퍼링크(간선)

## 2. 그래프를 분류하는 기준 💡

| 분류 기준 | 종류 | 설명 |
|---|---|---|
| 방향성 | 무방향 그래프 | 친구 관계처럼 양방향 (A↔B) |
|  | 유방향 그래프 | 팔로우처럼 단방향 (A→B) |
| 가중치 | 가중치 그래프 | 간선에 비용이 있음 (서울→부산: 400km) |
| 순환 여부 | 사이클(Cycle) | 출발점으로 돌아올 수 있는 경로 존재 |

## 3. 그래프를 저장하는 두 가지 방법 ⚠️

### ① 인접 행렬 (Adjacency Matrix)
2차원 배열(표)로 연결 여부를 저장

|   | A | B | C |
|---|---|---|---|
| A | 0 | 1 | 0 |
| B | 1 | 0 | 1 |
| C | 0 | 1 | 0 |

| 장점 | 단점 |
|---|---|
| 연결 확인 O(1) | 메모리 낭비 (희소 그래프에서) |

### ② 인접 리스트 (Adjacency List)
각 정점마다 연결된 정점들을 리스트로 관리

```python
graph = {
    'A': ['B'],
    'B': ['A', 'C'],
    'C': ['B']
}
```
| 장점 | 단점 |
|---|---|
| 메모리 효율적 | 연결 확인 O(V) |

### 4. 그래프를 탐색하는 두 가지 방법

| 방법 | 철학 | 자료구조 |
|---|---|---|
| DFS (깊이 우선) | "끝까지 가보자!" | 스택 / 재귀 |
| BFS (너비 우선) | "주변부터 확인!" | 큐 |

💡 핵심 포인트
**트리(Tree)**는 그래프의 일종이지만, 사이클이 없고 모든 노드가 연결된 특별한 형태입니다.

❓ 확인 질문
내비게이션 앱: 수만 개의 교차로(정점), 한 교차로당 3~4개의 길(간선)만 연결

👉 인접 행렬 vs 인접 리스트 중 무엇이 메모리 효율적일까요?

정답: 간선이 적으므로 → 인접 리스트 ✅
