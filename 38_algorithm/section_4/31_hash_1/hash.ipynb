```python
class Dict:
    def __init__(self):
        self.item = [None] * 8
    
    def put(self, key, value):
        index = hash(key) % len(self.item)
        self.item[index] = value
    
    def get(self, key):
        return self.item[hash(key) % len(self.item)]
```

```python
dict = Dict()
```

```python
dict.put("apple", "사과")
dict.put("orange", "오렌지")
```

```python
dict.item
```

```python
dict.get("orange")
```

```python
class LinkedTuple:
    def __init__(self):
        self.item = []
    
    def put(self, key, value):
        self.item.append((key, value))
    
    def get(self, key):
        for k, v in self.item:
            if k == key:
                return v
```

```python
class LinkedDict:
    def __init__(self):
        self.item = []
        for _ in range(8):
            self.item.append(LinkedTuple())

    def put(self, key, value):
        index = hash(key) % len(self.item)
        self.item[index].put(key, value)

    def get(self, key):
        index = hash(key) % len(self.item)
        return self.item[index].get(key)
```

```python
linked_dict = LinkedDict()
```
