```python
a, b = map(int, input().split())

if a > b:
    print(">")

if a < b:
    print("<")

if a == b:
    print("==")
```

```python
num = int(input())

if 100 >= num >= 90:
    print("A")

elif 89 >= num >= 80:
    print("B")

elif 79 >= num >= 70:
    print("C")

elif 69 >= num >= 60:
    print("D")
    
else:
    print("F")


```
